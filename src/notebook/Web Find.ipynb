```python
x = page.find('<div id="mw-content-text" lang="en" dir="ltr" class="mw-content-ltr"><h2><span class="mw-headline" id="Problem">Problem</span></h2>')
shift = 135
strEnd = -1;
for i in range(x+shift, len(page)):
    if (page[i] == '?'):
        strEnd = i + 1
        break

print(page[x+shift:strEnd])
```
